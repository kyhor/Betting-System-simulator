```python
game_edge = 0.96

from random import *
import numpy

def wdlGame (prob, Mbet):
    ret = [game_edge/p for p in prob]
    Mwdist = float(randint(0, 100))/100
    remain = int (100 - Mwdist*100)
    Mddist = float(randint(0, remain))/100
    Mldist = 1 - Mwdist - Mddist
    game_result = numpy.random.choice(['w','d','l'], p=[prob[0],prob[1],prob[2]])
    if game_result == 'w':
        profit = Mbet*( Mldist + Mddist - Mwdist*(ret[0]-1))
    elif game_result == 'd':
        profit = Mbet*( Mwdist + Mldist - Mddist*(ret[1]-1))
    elif game_result == 'l':
        profit = Mbet*( Mwdist + Mddist - Mldist*(ret[2]-1))
    print 'profit: ',profit 
    return profit
    

def probGenerator():
    pw = float(randint(1,949))/1000
    remain = int (949 - pw*1000)
    pd = float(randint(1, remain))/1000
    pl = 1 - pw - pd
    return [pw,pd,pl]

def bigsmall(binprob,Mbet):
    ret = [game_edge/p for p in binprob]
    Mwdist = float(randint(0, 100))/100
    Mldist = 1 - Mwdist
    game_result = numpy.random.choice(['w','l'], p=[prob[0],prob[1]])
    if game_result == 'w':
        profit = Mbet*( Mldist - Mwdist*(ret[0]-1))
    elif game_result == 'l':
        profit = Mbet*( Mwdist - Mldist*(ret[1]-1))
    print 'profit: ',profit 
    return profit
    
    
def multipossiblegames(n_poss,n_prob,Mbet):
    ret = [game_edge/p for p in n_prob]
    mdist = list()
    results = list()
    results.append(0)
    M0 = float(randint(0, 100))/100
    sums = M0
    for i in range(1,n_poss):
        results.append(i)
        remain = int (100 - sums*100)
        mi = float(randint(0, remain))/100
        mdist.append(mi)
        sums += mi
    game_result = numpy.random.choice(results, p=n_prob)
    gain = 0
    for i in range(n_poss):
        if i != game_result:
            gain += Mbet*mdist[i]
    profit = gain - Mbet*(mdist[game_result]-1)
    return profit
    
def cleverbet(prob):
    newdist = list()
    for p in prob:
        if p >= 0.96:
            newdist.append(p)
        else:
            newdist.append(0.0)
    return newdist
    
def wdlGame2 (prob, Mbet, Mdist):
    ret = [game_edge/p for p in prob]
    Mwdist = Mdist[0]
    Mddist = Mdist[1]
    Mldist = Mdist[2]
    game_result = numpy.random.choice(['w','d','l'], p=[prob[0],prob[1],prob[2]])
    if game_result == 'w':
        profit = Mbet*( Mldist + Mddist - Mwdist*(ret[0]-1))
    elif game_result == 'd':
        profit = Mbet*( Mwdist + Mldist - Mddist*(ret[1]-1))
    elif game_result == 'l':
        profit = Mbet*( Mwdist + Mddist - Mldist*(ret[2]-1))
    print 'profit: ',profit 
    return profit
```

```python
storage_cash = 10000000

win = 0
lose = 0
rounds = 1000
totalprof = 0
for i in range(rounds):
    prob = probGenerator()
    Mb = randint(3000000, 5000000)
    print 'round ',i,' Mb:', Mb
    profit = wdlGame2(prob,Mb,cleverbet(prob))
    if profit > 0:
        win += 1
    elif profit < 0:
        lose += 1
    storage_cash += profit
    print 'round ',i,' cash:', storage_cash
    print '----------------------------------------'
#print 'win rate:', float(win)/rounds
print 'win:', float(win)
print 'lose:', float(lose)
      
```
